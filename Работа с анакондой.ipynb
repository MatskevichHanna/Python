```python
from sympy import *
```

**Инициализация символов

```python
x = Symbol('x')
function = (sin(x)/2 + cos(x)/2) ** 2 * sin(x)
function
```

**Находим периодичность функции

```python
period = periodicity(function, x)
period
```

**Рассматриваем отдельный период

```python
period_interval = (-period/2, period/2)
period_interval
```

**Находим корни

```python
roots = solve(function, x)
print(roots)
```

**Находим интервалы, где функция возрастает и где убывает

```python
diff_func = sorted(solve(diff(function), x))
diff_func.insert(0, period_interval[0])
diff_func.append(period_interval[1])
int_symb = []
for i in range(len(diff_func) - 1):
    if(diff(function).subs(x, (diff_func[i] + diff_func[i + 1])/2)) > 0:
        int_symb.append('возрастает')
    else:
        int_symb.append('убывает')
    print(f'На интервале {(diff_func[i].evalf(3), diff_func[i + 1].evalf(3))} функция {int_symb[-1]}')
```

**Строим график функции

```python
graf = plot(function, xlim = period_interval)
```

**Находим вершины

```python
points = []
for i in diff_func:
    points.append((function.subs(x, i), i))
print(f'Верхняя вершина в координате - {(max(points)[1].evalf(3), max(points)[0].evalf(3))}')
print(f'Верхняя вершина в координате - {(min(points)[1].evalf(3), min(points)[0].evalf(3))}')
```

**Определяем промежутки, на которых функция больше 0 и меньше 0

```python
roots = sorted(roots)
roots.insert(0, period_interval[0])
int_symb = []
for i in range(len(roots) - 1):
    if(function.subs(x, (roots[i] + roots[i + 1])/2)) > 0:
        int_symb.append('больше нуля')
    else:
        int_symb.append('меньше нуля')
    print(f'На интервале {(roots[i].evalf(3), roots[i + 1].evalf(3))} функция {int_symb[-1]}')
```
